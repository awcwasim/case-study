**Feature Engineering**

```python
num_value = [name for name in student_df.columns if student_df[name].dtype in ['int64', 'float64']]
cat_value = [name for name in student_df.columns if student_df[name].dtype == 'object']
```

```python
cat_value
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in list(cat_value):
    student_df[i]=le.fit_transform(student_df[i])
```

```python
student_df.info()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(student_df.corr().abs(), annot = True)
```

```python
x = student_df.drop('G3',axis=1)
y = student_df['G3']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
lin_model.fit(x_train,y_train)
```

**ML**

```python
from sklearn.metrics import r2_score
y_pred = lin_model.predict(x_test)
print(r2_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_model = DecisionTreeRegressor(min_samples_split= 3, max_depth=4)
dec_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = dec_model.predict(x_test)
print(r2_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
ran_model = RandomForestClassifier(n_estimators = 150, min_samples_split= 4)
ran_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = ran_model.predict(x_test)
print(r2_score(y_test, y_pred))
```

```python
from sklearn.svm import SVR
svm_model = SVR(kernel= 'linear', C= 2)
svm_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = svm_model.predict(x_test)
print(r2_score(y_test, y_pred))
```

Random forest gives better prediction
